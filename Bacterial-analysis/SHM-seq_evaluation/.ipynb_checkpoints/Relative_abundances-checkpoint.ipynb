### Explore relative abundances SHM-Seq vs qPCR (ASF) and for SPF

Input files:<br>
ASF Bac pipeline output file (tsv): QC16S_ASF_$*$Bacstdata.tsv<br>
SPF Bac pipeline output file (tsv): S1_$*$Bacstdata.tsv


```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter, defaultdict
import seaborn as sns
import pickle
import matplotlib as mpl
from sklearn import preprocessing
from scipy.stats import linregress, spearmanr, pearsonr
import warnings
warnings.filterwarnings('ignore')
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams["font.size"] = 18
```

```python
file_names = {'QC16S_ASF_C1_S1':['ASF', 'C1', 'Treatment'],
             'QC16S_ASF_D1_S2':['ASF', 'D1', 'Treatment'],
             'QC16S_ASF_E1_S3':['ASF', 'E1', 'Treatment'],
             'S1_C2':['WT','C2', 'Treatment'],
             'S1_D2':['WT','D2', 'Treatment'],
             'S1_E2':['WT','E2', 'Treatment']}

asf_short = {'Clostridium,Clostridiaceae,Clostridiales,Clostridia,Firmicutes,Bacteria':'ASF356',
             'Eubacterium,Eubacteriaceae,Clostridiales,Clostridia,Firmicutes,Bacteria':'ASF492',
             'Lactobacillus,Lactobacillaceae,Lactobacillales,Bacilli,Firmicutes,Bacteria':'ASF360',
             "Ligilactobacillus,Lactobacillaceae,Lactobacillales,Bacilli,Firmicutes,Bacteria":'ASF361',
             'Mucispirillum,Deferribacteraceae,Deferribacterales,Deferribacteres,Deferribacteres,Bacteria':'ASF457',
             'Parabacteroides,Tannerellaceae,Bacteroidales,Bacteroidia,Bacteroidetes,Bacteria':'ASF519',
             'Schaedlerella,Lachnospiraceae,Clostridiales,Clostridia,Firmicutes,Bacteria':'ASF502',
            'Unassigned,Unassigned,Unassigned,Unassigned,Firmicutes,Bacteria':'ASF500',
            'Unassigned,Lactobacillaceae,Lactobacillales,Bacilli,Firmicutes,Bacteria':'Unassigned',
            'Unassigned,Unassigned,Clostridiales,Clostridia,Firmicutes,Bacteria':'Unassigned',  
         'Unassigned,Unassigned,Unassigned,Unassigned,Unassigned,Bacteria':'Unassigned'  }
```

```python
# data from table 2 (Spatial Distribution and Stability of the Eight Microbial Species of the Altered Schaedler Flora in the Mouse Gastrointestinal Tract)

# Middle
external_data = {'ASF356':[547.6, 53.9, 46.5],
                 'ASF361':[1.0, 1.0, 1.0],
           'ASF457':[1060.7, 213.9, 190.7],
           'ASF492':[424.0, 10.3, 42.5],
           'ASF500':[81.0, 6.1, 8.7],
           'ASF502':[2388.8, 209.1, 289.1],
           'ASF519':[3597.2, 1357.4, 296.2],
            'ASF360':[0, 0, 0]}

external_df = pd.DataFrame(external_data)
external_df = external_df.T
external_df.columns = ['m1', 'm2', 'm3']
for mouse in external_df.columns:
    colsum = external_df[mouse].sum()
    external_df[mouse + '.rel.abund'] = external_df[mouse] / colsum *100
external_df.reset_index(inplace=True)
cols = [i for i in external_df.columns if i.endswith('.rel.abund')]
external_df = pd.melt(external_df, id_vars=['index'], value_vars=cols)
external_df['condition'] = 'external'
external_df.rename(columns={'index':'genus'}, inplace=True)
print(external_df.head())
```

# tsv files

```python
# Read log files

path = ''

taxa_order = ['genus', 'family', 'order', 'class', 'phylum']

tsv_l = []

for file in os.listdir(path):
    if file.endswith(".tsv"):
        name = file.split('/')[-1].split('_Bac')[0]
        
        if name in file_names:
            #print(name)
            df = pd.read_csv(os.path.join(path, file), sep='\t')
            df.drop(columns='Unnamed: 0', inplace=True)

            if name.startswith('QC16S_ASF'):
                df['genus'] = df['taxa_order'].replace(asf_short)
            else:
                df['genus'] = df['taxa_order'].str.split(',').str[0]

            df = df[df['genus'] != 'Unassigned']
            df = df.groupby('genus').sum()
            df['rel.abund'] = df['fastq'] / sum(df['fastq'].tolist()) *100
            df['index'] = name  

            tsv_l.append(df)
        
f = pd.concat(tsv_l)
f['xy'] = f['index'].map(file_names)
f[['mouse', 'well', 'Treat']] = pd.DataFrame(f['xy'].tolist(), index=f.index)
f.reset_index(inplace=True)

print(f.head())
```

## ASF

```python
# Plot scatter plot on genus level
pd.set_option('max_colwidth', None)
fig = plt.figure(figsize=(10,5))

# for a 1x2 subplot
axes = []
for row_idx in range(0,1):
    for col_idx in range(0,1):
        ax = plt.subplot2grid((1, 1), (row_idx, col_idx))
        axes.append(ax)
        
x = 'spatial'
y = 'external'
hue = 'genus'  
ax = axes[0]
palette = {'ASF356':'pink',
          'ASF360':'plum', 
           'ASF361':'fuchsia', 
           'ASF457':'blueviolet', 
           'ASF492':'midnightblue', 
           'ASF500':'cornflowerblue',
           'ASF502':'skyblue',
           'ASF519':'teal'}

data = f[f['mouse'] == 'ASF']
data = data[data['Treat'] == 'Treatment']
data = pd.melt(data, id_vars=['genus'], value_vars=['rel.abund'])
data.sort_values(by='value', ascending=False, inplace=True)
data['condition'] = 'spatial'

data = data.loc[:,['genus', 'value']]
data.columns = ['genus', 'spatial']

external_df2 = external_df.loc[:,['genus', 'value']]
external_df2.columns = ['genus', 'external']
external_df2.sort_values(by=['genus', 'external'], inplace=True)
external_df2.reset_index(inplace=True)

data = data[data['genus'].isin(list(set(external_df2['genus'].tolist())))]
data.sort_values(by=['genus', 'spatial'], inplace=True)
data.reset_index(inplace=True)

dataDF = pd.concat([data,external_df2], axis=1).iloc[:,[1, 2, 5]]
dataDF[["spatial", "external"]] = dataDF[["spatial", "external"]].apply(pd.to_numeric)

sns.regplot(x=x, y=y, data=dataDF, ax=ax, scatter_kws={'rasterized':True, 's':1}, line_kws={"color": "lightgrey"})
sns.scatterplot(x=x, y=y, data=dataDF, ax=ax, hue=hue, palette=palette)
ax.set_xlabel('Spatial host-microbiome sequencing \n Relative abundance (%)')
ax.set_ylabel('External qPCR \nRelative abundance (%)')


# Pearson's r
pear, pp = pearsonr(dataDF[x], dataDF[y])
print("Pearson's r: " + str(round(pear, 3)) + ' pvalue: ' + str(pp))
ax.text(0.02, 0.94, "Pearson's r: " + str(round(pear, 2)), transform=ax.transAxes, fontsize=15)

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.tight_layout()
```

```python
# Plot stacked barplot on genus level
pd.set_option('max_colwidth', None)
fig = plt.figure(figsize=(20,10))

# for a 1x1 subplot
axes = []
for row_idx in range(0,1):
    for col_idx in range(0,1):
        ax = plt.subplot2grid((1, 1), (row_idx, col_idx))
        axes.append(ax)
yourlist = ['rep1','rep2','rep3']
ntimes = 8 # number of times you repeat each item

# Spatial data
sg = f[f['mouse'] == 'ASF']
sg = sg[sg['Treat'] == 'Treatment']
sg = pd.melt(sg, id_vars=['genus'], value_vars=['rel.abund'])
sg['z'] = [i for i in yourlist for _ in range(ntimes)]
sg['var_z'] = 'spatial_' + sg['z']
sg.drop(columns=['variable', 'z'], inplace=True)

# External
ext = external_df.copy()
ext['z'] = [i for i in yourlist for _ in range(ntimes)]
ext['var_z'] = 'external_' + ext['z']
ext.drop(columns=['variable', 'z', 'condition'], inplace=True)

# Merge
g = pd.concat([sg, ext])
g = g.pivot(index='var_z', columns='genus')['value']
g.reset_index(inplace=True)
g = g.rename_axis(None, axis=1)
g.set_index('var_z', inplace=True)

# plot data in stack manner of bar type
g.loc[['spatial_rep1', 'spatial_rep2', 'spatial_rep3','external_rep1', 'external_rep2', 'external_rep3']].reset_index().plot(x='var_z', kind='bar', stacked=True, color=palette, ax=ax)
ax.set_xlabel('')
ax.set_ylabel('Relative abundance (%)')
ax.legend(bbox_to_anchor=(1.0, 0.75))
ax.set_xticklabels(['SHM-seq\nReplicate 1', 'SHM-seq\nReplicate 2','SHM-seq\nReplicate 3',
                   'External qPCR\nReplicate 1', 'External qPCR\nReplicate 2','External qPCR\nReplicate 3'], rotation=0)

plt.tight_layout()
```

 ## WT

```python
# Plot stacked barplot on genus level
pd.set_option('max_colwidth', None)
fig = plt.figure(figsize=(20,10))

# for a 1x1 subplot
axes = []
for row_idx in range(0,1):
    for col_idx in range(0,1):
        ax = plt.subplot2grid((1, 1), (row_idx, col_idx))
        axes.append(ax)
wt_palette = {'Other':'grey', 
              'Duncaniella':'lightcoral', 
              'Turicibacter':'maroon',
              'Muribaculum':'peru', 
              'Ruminococcus':'tan', 
              'Clostridium':'orange', 
              'Lactobacillus':'gold',
              'Lachnoclostridium':'yellowgreen', 
              'Butyrivibrio':'palegreen', 
              'Oscillibacter':'lightseagreen', 
              'Blautia':'powderblue'}
yourlist = ['rep1','rep2','rep3']
ntimes = 39 # number of times you repeat each item

# Spatial data
wg = f[f['mouse'] == 'WT']
wg = wg[wg['Treat'] == 'Treatment']
wg = pd.melt(wg, id_vars=['genus'], value_vars=['rel.abund'])
wg = wg[~wg['genus'].isin(['Unassigned', 'Acetobacterium', 'Lacrimispora'])]
wg['z'] = [i for i in yourlist for _ in range(ntimes)]
wg.drop(columns=['variable'], inplace=True)
wg.sort_values(by='value', inplace=True, ascending=False)
wg = wg.groupby('z').head(10)
wg_tot = wg.groupby('z').sum()
wg_d = {}
for ind, row in wg_tot.iterrows():
    wg_d[ind] = ['Other', 100-row['value']]
wg_new = pd.DataFrame.from_dict(wg_d, columns=['genus', 'value'], orient='index')
# Add others
wg = pd.concat([wg_new, wg.set_index('z')])
wg.reset_index(inplace=True)
wg = wg.pivot(index='index', columns='genus')['value']
wg.reset_index(inplace=True)
wg = wg.rename_axis(None, axis=1)
wg.set_index('index', inplace=True)
cols = ['Blautia', 'Butyrivibrio', 'Clostridium', 'Duncaniella', 'Lachnoclostridium', 'Lactobacillus', 'Muribaculum', 'Oscillibacter', 'Ruminococcus', 'Turicibacter', 'Other']
wg = wg[cols]

# plot data in stack manner of bar type
wg.reset_index().plot(x='index', kind='bar', stacked=True, ax=ax, color=wt_palette)
ax.set_xlabel('')
ax.set_ylabel('Relative abundance (%)')
ax.legend(bbox_to_anchor=(1, 0.75))
ax.set_xticklabels(['Replicate 1', 'Replicate 2','Replicate 3'], rotation=0)

plt.tight_layout()
```
