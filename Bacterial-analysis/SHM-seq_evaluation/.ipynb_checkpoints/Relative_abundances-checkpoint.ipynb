### Explore relative abundances SHM-Seq vs qPCR (ASF) and for SPF

Input files:<br>
SPF Bac pipeline output file (tsv): S1_$*$Bacstdata.tsv


```python
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter, defaultdict
import seaborn as sns
import pickle
import matplotlib as mpl
from sklearn import preprocessing
from scipy.stats import linregress, spearmanr, pearsonr
import warnings
warnings.filterwarnings('ignore')
import matplotlib
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
matplotlib.rcParams["font.size"] = 18
```

```python
file_names = {'S1_C2':['WT','C2', 'Treatment'],
             'S1_D2':['WT','D2', 'Treatment'],
             'S1_E2':['WT','E2', 'Treatment']}
```

# tsv files

```python
# Read log files

path = ''

taxa_order = ['genus', 'family', 'order', 'class', 'phylum']

tsv_l = []

for file in os.listdir(path):
    if file.endswith(".tsv"):
        name = file.split('/')[-1].split('_Bac')[0]
        
        if name in file_names:
            #print(name)
            df = pd.read_csv(os.path.join(path, file), sep='\t')
            df.drop(columns='Unnamed: 0', inplace=True)

            df['genus'] = df['taxa_order'].str.split(',').str[0]

            df = df[df['genus'] != 'Unassigned']
            df = df.groupby('genus').sum()
            df['rel.abund'] = df['fastq'] / sum(df['fastq'].tolist()) *100
            df['index'] = name  

            tsv_l.append(df)
        
f = pd.concat(tsv_l)
f['xy'] = f['index'].map(file_names)
f[['mouse', 'well', 'Treat']] = pd.DataFrame(f['xy'].tolist(), index=f.index)
f.reset_index(inplace=True)

print(f.head())
```

 ## WT

```python
# Plot stacked barplot on genus level
pd.set_option('max_colwidth', None)
fig = plt.figure(figsize=(20,10))

# for a 1x1 subplot
axes = []
for row_idx in range(0,1):
    for col_idx in range(0,1):
        ax = plt.subplot2grid((1, 1), (row_idx, col_idx))
        axes.append(ax)
wt_palette = {'Other':'grey', 
              'Duncaniella':'lightcoral', 
              'Turicibacter':'maroon',
              'Muribaculum':'peru', 
              'Ruminococcus':'tan', 
              'Clostridium':'orange', 
              'Lactobacillus':'gold',
              'Lachnoclostridium':'yellowgreen', 
              'Butyrivibrio':'palegreen', 
              'Oscillibacter':'lightseagreen', 
              'Blautia':'powderblue'}
yourlist = ['rep1','rep2','rep3']
ntimes = 39 # number of times you repeat each item

# Spatial data
wg = f[f['mouse'] == 'WT']
wg = wg[wg['Treat'] == 'Treatment']
wg = pd.melt(wg, id_vars=['genus'], value_vars=['rel.abund'])
wg = wg[~wg['genus'].isin(['Unassigned', 'Acetobacterium', 'Lacrimispora'])]
wg['z'] = [i for i in yourlist for _ in range(ntimes)]
wg.drop(columns=['variable'], inplace=True)
wg.sort_values(by='value', inplace=True, ascending=False)
wg = wg.groupby('z').head(10)
wg_tot = wg.groupby('z').sum()
wg_d = {}
for ind, row in wg_tot.iterrows():
    wg_d[ind] = ['Other', 100-row['value']]
wg_new = pd.DataFrame.from_dict(wg_d, columns=['genus', 'value'], orient='index')
# Add others
wg = pd.concat([wg_new, wg.set_index('z')])
wg.reset_index(inplace=True)
wg = wg.pivot(index='index', columns='genus')['value']
wg.reset_index(inplace=True)
wg = wg.rename_axis(None, axis=1)
wg.set_index('index', inplace=True)
cols = ['Blautia', 'Butyrivibrio', 'Clostridium', 'Duncaniella', 'Lachnoclostridium', 'Lactobacillus', 'Muribaculum', 'Oscillibacter', 'Ruminococcus', 'Turicibacter', 'Other']
wg = wg[cols]

# plot data in stack manner of bar type
wg.reset_index().plot(x='index', kind='bar', stacked=True, ax=ax, color=wt_palette)
ax.set_xlabel('')
ax.set_ylabel('Relative abundance (%)')
ax.legend(bbox_to_anchor=(1, 0.75))
ax.set_xticklabels(['Replicate 1', 'Replicate 2','Replicate 3'], rotation=0)

plt.tight_layout()
```
