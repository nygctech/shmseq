### Overlay of HE, spots and annotation.

Input files:<br>
Representative HE images: SPF(10015CN41_D2_HE.jpg), GF(10015CN72_C2_HE.jpg), ASF(10015CN26_D2_HE.jpg)<br>
Annotation files for the same representative samples. 

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
from PIL import Image
Image.MAX_IMAGE_PIXELS = 1000000000
from skimage import io
from skimage.color import rgb2gray
import warnings
warnings.filterwarnings('ignore')
```

```python
def resize_func(arr):
    """
    Resize image after ST well size.
    Returns x and y coordinates of the image fiited in a ST well, as well as the resized image. 
    """
    from skimage.transform import resize

    # Rescale
    image_rescaled = resize(arr, (arr.shape[0] // 10, arr.shape[1] // 10),anti_aliasing=True, preserve_range=True) #10

    xdim,ydim = image_rescaled.shape 

    testy = ydim/32-1
    testx = xdim/34-1

    x_indices = np.array([i*testy for i in range(1, 33)])
    y_indices = np.array([i*testx for i in range(1, 35)])

    x_coordinates = []
    y_coordinates = []
    for i in x_indices:
        for j in y_indices:
            x_coordinates.append(i)
            y_coordinates.append(j)

    return x_coordinates, y_coordinates, image_rescaled

from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, color_map, sort_colors=True, emptycols=0):

    inv_map = {v: k for k, v in color_map.items()}
    
    cell_width = 212
    cell_height = 22
    swatch_width = 10
    margin = 6

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in colors.items())
        names = [name for hsv, name in by_hsv]
    else:
        names = list(colors)

    n = len(names)
    ncols = 5 - emptycols
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + 2 * margin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-margin)/height)
    ax.set_xlim(0, cell_width * 5)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, inv_map[name], fontsize=8,
                horizontalalignment='left',
                verticalalignment='center')

        ax.add_patch(
            Rectangle(xy=(swatch_start_x, y-9), width=swatch_width,
                      height=18, facecolor=colors[name], edgecolor='0.7')
        )

    return fig
```

```python
rename_dict = {"muscularis mucosae and muscularis propria muscularis interna and peyer's patch":"mucosae and peyer's patch",
               "epithelium and peyer's patch":"peyer's patch",
                  'epithelium and lamina propria' : 'epithelium',
    'epithelium and lamina propria and mucosa' : 'epithelium and mucosa',
    'epithelium and lamina propria and mucosa and pellet': 'epithelium and mucosa and pellet',
    'epithelium and lamina propria and muscularis mucosae' : 'epithelium and mucosae',
    'epithelium and lamina propria and muscularis mucosae and muscularis propria muscularis externa and muscularis propria muscularis interna and submucosa all':'epithelium and muscle and submucosa',
    'epithelium and lamina propria and muscularis mucosae and submucosa all':'epithelium and mucosae and submucosa',
    "epithelium and lamina propria and peyer's patch" : "peyer's patch",
    'epithelium apex of crypt and lamina propria' : 'crypt apex',
    'epithelium base of crypt and epithelium mid crypt and lamina propria' : 'crypt base and mid',
    'epithelium base of crypt and lamina propria' : 'crypt base',
    'epithelium base of crypt and lamina propria and muscularis mucosae and muscularis propria muscularis externa and muscularis propria muscularis interna and submucosa all' : 'crypt base',
    'epithelium base of crypt and muscularis mucosae and muscularis propria muscularis interna': 'crypt base',
    'epithelium mid crypt and lamina propria' : 'crypt mid',
    'mucosa and pellet' : 'mucosa and pellet',
    'muscularis all and submucosa all' : 'muscle and submucosa',
    'muscularis mucosae and muscularis propria muscularis externa and muscularis propria muscularis interna' : 'muscle and submucosa',
    'muscularis mucosae and muscularis propria muscularis interna' : 'mucosae and interna',
    "muscularis mucosae and muscularis propria muscularis interna and peyer's patch" : "peyer's patch",
    "muscularis mucosae and peyer's patch" : "peyer's patch",
    'muscularis propria muscularis externa' : 'externa',
    'muscularis propria muscularis externa and muscularis propria muscularis interna':'externa and interna',
    'muscularis propria muscularis interna':'interna',
    'crypt apex and crypt mid':'crypt mid',
    'muscularis propria muscularis externa and muscularis propria muscularis interna and muscularis mucosae and submucosa all' : 'muscle and submucosa',
    'pellet':'pellet',
    "peyer's patch":"peyer's patch",
    'mucosa':'mucosa',
    'epithelium apex of crypt and mucosa':'crypt apex and mucosa',
    'epithelium apex of crypt and lamina propria and mucosa':'crypt apex and mucosa',
    "Rest": "rest"}

rename_value = list(set([v for k,v in rename_dict.items()]))
```

```python
# Load HE image - double checked that these files are identical to the ones on Fire Cloud Broad Institute
img_url = {'SPF':'10015CN41_D2_HE.jpg',
          'ASF':'10015CN26_D2_HE.jpg',
          'GF':'10015CN72_C2_HE.jpg'}

anno_paths = {'SPF':'10015CN41_D2_annotations.txt',
              'ASF':'10015CN26_D2_annotations.txt',
              'GF':'10015CN72_C2_annotations.txt'}

entities = {} 
for tissue_type in ['GF', 'SPF', 'ASF']:
    d = {}
    print(tissue_type)
    img = np.array(Image.open(img_url[tissue_type]))
    d['he_image'] = img
    
    # Read input image
    he_bw = rgb2gray(io.imread(img_url[tissue_type]))
    he_x, he_y, he_resize = resize_func(he_bw) # resize to ST measurements
    he_spots = pd.DataFrame({'x':he_x, 'y':he_y})
    d['he_resize'] = he_resize
    d['he_spots'] = he_spots
    
    # Read annotation
    ann = pd.read_csv(anno_paths[tissue_type], sep='\t')
    
    # To get xdim
    xdim,ydim = he_resize.shape

    testy = ydim/32
    testx = xdim/34

    x_indices = np.array([(i-1)*testy for i in ann['x']])
    y_indices = np.array([(i-1)*testx for i in ann['y']])

    # add correct pixel coordiantes
    ann['new_x'] = x_indices
    ann['new_y'] = y_indices
    
    # Check if ann includes incorrect annotation
    if ann[~ann['value'].isin(rename_value)].shape[0] != 0:
        ann['value'] = ann['value'].replace(rename_dict)
     
    d['annotation'] = ann
    
    entities[tissue_type] = d

```

```python
# Create common color map
import matplotlib.colors as mcolors
from random import seed, sample

annos_l = []
for k,v in entities.items():
    
    #Store annotation in list fro all tissue types
    for ll in v['annotation']['value'].tolist():
        annos_l.append(ll)
        
annos_l = list(set(annos_l))

# Randomly select colors
seed(10)

colors = sample(list(mcolors.CSS4_COLORS), len(annos_l))

selected_colors = {k:v for k, v in mcolors.CSS4_COLORS.items() if k in colors}
print(selected_colors)

color_map = {ann:colors[i] for i, ann in enumerate(annos_l)}
print(color_map)

num_plots = len(annos_l)
x = np.arange(num_plots)

for i in range(1, num_plots):
    plt.plot(x, i * x + 5 * i, color=colors[i])
```

```python
# initialize a figure
fig = plt.figure(figsize=(18, 18))

axes = []
for ax_idx in range(0,3):
    for i in range(3):
        ax = plt.subplot2grid((3, 3), (ax_idx, i))
        axes.append(ax)
       
for i, tissue_type in enumerate(list(img_url)):
    
    # visualize the tissue image
    ax = axes[i*3]
    ax.imshow(entities[tissue_type]['he_image'])
    ax.set_xticks([])
    ax.set_yticks([])
    
    # A, B and C
    if i == 0:
        ax.text(-0.1, 1,'a', transform=ax.transAxes, fontsize=25)
    elif i == 1:
        ax.text(-0.1, 1,'b', transform=ax.transAxes, fontsize=25)
    elif i == 2:
        ax.text(-0.1, 1,'c', transform=ax.transAxes, fontsize=25)
    

    # Visulalize spots and HE image
    ax = axes[i*3+1]
    sns.scatterplot(x='x', y='y', data=entities[tissue_type]['he_spots'], ax=ax)
    ax.imshow(entities[tissue_type]['he_resize'], cmap='gray')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticks([])
    ax.set_yticks([])

    # visualize the ST spots on top of the tissue image
    ax = axes[i*3+2]
    sns.scatterplot(x='new_x', y='new_y', hue='value', data=entities[tissue_type]['annotation'], ax=ax, palette=color_map)
    ax.legend(loc='center left', bbox_to_anchor=(1, 0.5),fontsize=13)
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticks([])
    ax.set_yticks([])
    ax.invert_yaxis()
    ax.imshow(entities[tissue_type]['he_resize'], cmap='gray')
    ax.get_legend().remove()
        
plt.show()
```

```python
plot_colortable(selected_colors, color_map)
plt.show()
```
