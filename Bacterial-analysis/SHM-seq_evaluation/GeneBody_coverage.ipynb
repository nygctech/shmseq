### Gene body coverage

Input files:<br>
ASF 16S rRNA gene sequences (fasta): downloaded from Ensembl Bacteria<br>
Gene body coverage files, output from RSeQC: asf$*$.geneBodyCoverage.txt<br>

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
# Load 16S rRNA reference

path = ''

fasta_f = 'asf_16S_rRNA_gene.fasta'

with open(os.path.join(path, fasta_f)) as file:
    lines = file.readlines()
    lines = [line.rstrip() for line in lines]

bac_d = {}
for i, line in enumerate(lines):
    if line.startswith('>'): # New bac
        name = line
        seq_l = []
    elif line == '':
        seq = ''.join(seq_l)
        bac_d[name] = seq
    else:
        seq_l.append(line)
```

```python
# Revere complement to 16S surface probe
surface_probe = 'GGATTAGATACCCTGGTAGTC'

exp_cap_pos_d = {}

for k,v in bac_d.items():
    
    for item in k.split(' '):
        if item.startswith('ASF'):
            name = item
            
    # find start and end region of expected captured region
    seq_end = (v.find(surface_probe) + len(surface_probe)) / len(v) *100
    seq_start = (v.find(surface_probe) + len(surface_probe) - 400) / len(v) *100
   
    exp_cap_pos_d[name] = [seq_start, seq_end]
    
```

```python
# Load genebody coverage txt files

gbc_d = {}

for file in os.listdir(path):

    if file.endswith("geneBodyCoverage.txt"):

        name = file.split('.')[0]
        
        gbc = pd.read_csv(os.path.join(path, file), sep='\t', index_col=0)
        gbc = gbc.T
        gbc.columns = [col.split('_')[2] for col in gbc.columns]
        
        for col in gbc.columns:
            col_sum = gbc[col].sum()
            gbc[col + '(%)'] = gbc[col] / col_sum * 100
        
        gbc.reset_index(inplace=True)
        gbc = pd.melt(gbc, id_vars=['index'], value_vars=['C1(%)', 'D1(%)', 'E1(%)'])
        gbc['variable'] = gbc['variable'].str[:-3]
        gbc_d[name] = gbc
```

```python
# Plot
# initialize a figure
fig = plt.figure(figsize=(30, 20))

# for a 2x3 subplot
axes = []
for row_idx in range(0,2):
    for col_idx in range(0,3):
        ax = plt.subplot2grid((2, 3), (row_idx, col_idx))
        axes.append(ax) 

x = 'index'
y = 'value'

reads_in_exp_region_ALL = []

for i, asf in enumerate(list(gbc_d)):
    ax = axes[i]
    
    sns.lineplot(x=x, y=y, data=gbc_d[asf], ax=ax)

    ax.set_ylabel('Coverage (%)', fontsize=15)
    ax.set_title(asf.upper(), fontsize=20)
    ### X ticks
    ax.set_xlabel("Gene body percentile (5'->3')", fontsize=15)
    ax.tick_params(axis='both', which='major', labelsize=14)
    
    labels = [item.get_text() for item in ax.get_xticklabels()]
    lk = 0
    l = [0, 9, 19, 29, 39, 49, 59, 69, 79, 89, 99]
    for label in ax.xaxis.get_ticklabels():
        if lk in l:
            label.set_visible(True)
        else:
            label.set_visible(False)
        lk += 1
    ###
    
    ax.axvline(x=exp_cap_pos_d[asf.upper()][0], color='red', ls=':')
    ax.axvline(x=exp_cap_pos_d[asf.upper()][1], color='red', ls=':')
    
    reads_in_exp_region = []
    for sample in ['C1', 'D1', 'E1']:
        df = gbc_d[asf][gbc_d[asf]['variable'] == sample]
        
        df['index'] = pd.to_numeric(df['index'])
        
        df = df[(df['index'] >= exp_cap_pos_d[asf.upper()][0]) &
               (df['index'] <= exp_cap_pos_d[asf.upper()][1])]
        
        reads_in_exp_region.append(df['value'].sum())
        reads_in_exp_region_ALL.append(df['value'].sum())
        
    ax.text(0.63, 0.90, "Reads within expected \ncapture region: " + str(round(mean(reads_in_exp_region), 0)) + '%', transform=ax.transAxes, fontsize=15)
```
