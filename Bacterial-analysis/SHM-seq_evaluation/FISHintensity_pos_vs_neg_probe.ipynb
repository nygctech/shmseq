```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
Image.MAX_IMAGE_PIXELS = None
from skimage.morphology import closing, square, remove_small_objects
from skimage.color import rgb2gray
import warnings
warnings.filterwarnings('ignore')
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def remove_background(url, light_limit, dark_limit):
    """
    Remove background of input Cy3 image
    """
    from scipy import ndimage as ndi
    from skimage import io, img_as_ubyte, filters, morphology
    
    # Read input image
    img = rgb2gray(io.imread(url))
    # Find edges in image using the Sobel filter.
    sobel = filters.sobel(img)
    # Multi-dimensional Gaussian filter.
    blurred = filters.gaussian(sobel, sigma=2.0)
    
    ### find light spots - might need change of limit
    light_spots = np.array((img > light_limit).nonzero()).T
    
    ### find dark spots - might need change of limit
    dark_spots = np.array((img < dark_limit).nonzero()).T
    
    # Masking
    bool_mask = np.zeros(img.shape, dtype=np.bool)
    bool_mask[tuple(light_spots.T)] = True
    bool_mask[tuple(dark_spots.T)] = True
    seed_mask, num_seeds = ndi.label(bool_mask)

    # Watershedding
    ws = morphology.watershed(blurred, seed_mask)
    
    # Set background and have a look
    background = max(set(ws.ravel()), key=lambda g: np.sum(ws == g))
    background_mask = (ws == background)
    plt.imshow(~background_mask)
    plt.title('background mask:' + url.split('/')[-1])
    
    # Remove background
    cleaned = img * ~background_mask

    return cleaned

def preprocess_label_map(lbl_img):
    '''
    Parameters:
    ----------
    lbls: ndarray
        integer array where each pixel value corresponds to its assigned class.

    Returns:
    ----------
    cleaned: ndarray
        integer array with gaps, edge artifacts, and small objects removed.

    '''
    cleaned = closing(lbl_img, square(3))			# Fills in holes in foreground (nonzero pixels)

    return cleaned

def resize(arr):
    """
    Resize image after ST well size.
    Returns x and y coordinates of the image fiited in a ST well, as well as the resized image. 
    """
    from skimage.transform import resize

    # Rescale
    image_rescaled = resize(arr, (arr.shape[0] // 10, arr.shape[1] // 10),anti_aliasing=True, preserve_range=True) #10
    
    xdim,ydim = image_rescaled.shape

    testy = ydim/32-1
    testx = xdim/34-1

    x_indices = np.array([i*testy for i in range(1, 33)])
    y_indices = np.array([i*testx for i in range(1, 35)])

    x_coordinates = []
    y_coordinates = []
    for i in x_indices:
        for j in y_indices:
            x_coordinates.append(i)
            y_coordinates.append(j)

    return x_coordinates, y_coordinates, image_rescaled

def make_square_around_spot(ann, image_rescaled):
    """
    Makes a square around a ST spot by taking in ST coordinates. 
    Returns the df (ann) with x any max and min values of each spot.
    
    numpy array shape
    y
    y
    y
    .
    .
    .
    x x x x ...

    row
    row
    row
    .
    .
    .
       column column column ...

    To get a submatrix from a 2D array, array[index(row):index(row)+1, index(column):index(column)+1] 
    array.shape give (y, x), i.e. (rows, columns)
    """
    # Make a square area of each spot
    xminmax = []
    for i in ann['new_x'].tolist():
        xminmax.append((float(i)-xdim/(image_rescaled.shape[0]/5.5),float(i)+xdim/(image_rescaled.shape[0]/5.5))) #10

    yminmax = []
    for i in ann['new_y'].tolist():
        yminmax.append((float(i)-ydim/(image_rescaled.shape[0]/5.5),float(i)+ydim/(image_rescaled.shape[0]/5.5))) #12

    df1 = pd.DataFrame(xminmax, columns = ['xmin', 'xmax'])
    df2 = pd.DataFrame(yminmax, columns = ['ymin', 'ymax'])

    results = pd.concat([ann, df1, df2], axis=1, sort=False )
    results[['xmin', 'xmax', 'ymin', 'ymax']] = results[['xmin', 'xmax', 'ymin', 'ymax']].astype(int)
    
    return results

def intensity_submatrix(row, img):
    """
    Calculates intensity of a submatrix based on x and y min and max values.
    Returns the sum intensity of the submatrix.
    
    #print(image_rescaled[498:520, 408:428].sum())
    #print(image_rescaled.max(), image_rescaled.min())
    #print(image_rescaled[498:520, 408:428])
    
    """
    import math

    gfg = np.count_nonzero(img[int(row['ymin']):int(row['ymax']), int(row['xmin']):int(row['xmax'])])
    ll = img[int(row['ymin']):int(row['ymax']), int(row['xmin']):int(row['xmax'])].sum()# / gfg
    
    if (math.isinf(ll)) or (math.isnan(ll)):
        return 0
    else:
        return ll
```

```python

path = ''
ann_path = '' 

d = {}
names = ['pos', 'neg']
class_names = {0: 'Other', 63: 'BG', 127:'Tissue', 191:'Fiber', 255:'Bac'}

count = 0

for i, sample in enumerate(['63x_C2.jpg', '63x_D2.jpg']):
    
    well = sample.split('_')[1].split('.')[0]
    
    if i == 0:
        when = '63x_' + well + '_Simple Segmentation.tif'
        wh = 4
    elif i == 1:
        when = '63x_' + well + '_Simple Segmentation.tif'
        wh = 3
    
    tif = np.array(Image.open(os.path.join(path, when)).convert('L'))
    cy3 = np.array(Image.open(os.path.join(path, sample)))

    ## Perform the same preprocessing that is applied to class maps prior to cell quantification for Splotch ##
    cleaned = preprocess_label_map(tif)
    
    cleaned[(cleaned != 63) & (cleaned != 127) & (cleaned != 191) & (cleaned != 255)] = 0 
    clas = len(np.unique(np.array(cleaned)))
    
    # check how many ct channels in this sample
    if (clas < 4):
        indicator = np.setdiff1d(['0','63', '127', '191', '255'],[str(i) for i in np.unique(np.array(cleaned))])
        clas = clas + len(indicator)
        for j in range(len(indicator)):
            cleaned[j] = float(indicator[j])

    t = np.unique(np.array(cleaned))[wh]

    # subset per ct
    cells = cleaned == t

    cells = remove_small_objects(cells, min_size=5)
    v = cells

    # To get spots under tissue
    cy3_x, cy3_y, cy3_rescaled = resize(cy3) 
        
    # Read in annotation file
    ann = pd.read_csv(os.path.join(ann_path, date + '_' + well + '_annotations.txt'), sep ="\t")

    # Select pixels containing region 
    rt_px = cy3*v

    # find x, y coordinates in new image
    x_coordinates, y_coordinates, image_rescaled = resize(rt_px)

    # To get xdim
    xdim,ydim = image_rescaled.shape  
    
    testy = ydim/32
    testx = xdim/34

    x_indices = np.array([(i-1)*testy for i in ann['x']])
    y_indices = np.array([(i-1)*testx for i in ann['y']])

    x_coordinates = []
    y_coordinates = []
    for k in x_indices:
        for l in y_indices:
            x_coordinates.append(k)
            y_coordinates.append(l) 

    # add correct pixel coordiantes
    ann['new_x'] = x_indices
    ann['new_y'] = y_indices

    # if ann label "Unknown" rename it to Background
    ann['value'] = ann['value'].replace({'Unknown':'background'})

    # rename regions to make sure they survive tf handling
    new_names = [{x:x.lower().replace(" - "," ").replace(";"," and ")} for x in set(ann['value'])]
    vals = [v for v in new_names for k,v in v.items()]
    keys = [v for v in new_names for v,k in v.items()]
    ann['value'] = ann['value'].replace({keys[i]: vals[i] for i in range(len(vals))})

    # Make a circle area of each spot
    results_round = make_square_around_spot(ann, image_rescaled)

    # Calculate intensity of subimages based on index values in submatrices in df
    results_round['subintensity'] = results_round.apply(lambda row: intensity_submatrix(row, image_rescaled), axis=1)
    results_round[['subintensity']] = results_round[['subintensity']].astype(int)
    
    # Store region intensity df
    d[names[i]] = results_round
    
```
