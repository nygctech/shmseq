### Evaluate bacterial count per annotated area

Input file:<br>
Anndata fil GF / SPF: anndata_hm_norm_all_de_genes_n_Feb2022.h5ad<br>

```python
from __future__ import absolute_import, division, print_function

import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import colors as mcolors
import seaborn as sns
import scanpy as sc
import anndata
import sklearn
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
matplotlib.rcParams.update({'axes.linewidth': 0.25,
                            'xtick.major.size': 1,
                            'xtick.major.width': 0.25,
                            'ytick.major.size': 1,
                            'ytick.major.width': 0.25,
                            'pdf.fonttype': 42,
                            'font.sans-serif': 'Arial',
                            'patch.edgecolor' : 'black'})
sc.set_figure_params(scanpy=True, dpi=80, dpi_save=600, frameon=True, vector_friendly=True, fontsize=15, figsize=None, color_map=None, format='pdf', facecolor=None, transparent=False, ipython_format='png2x')
```

```python
# Read anndata file

tx_order = ['genus', 'family', 'order', 'class', 'phylum']
bac_regions = ['crypt apex and crypt mid', 'crypt apex and mucosa', 'crypt base', 'crypt mid', 'mucosa and pellet', 'pellet']

####### WEIGHTED MEAN #######

path = ''
fa = sc.read_h5ad(os.path.join(path, 'anndata_hm_norm_all_de_genes_n_Feb2022.h5ad'))
fa.obs['Age'] = '6w'
fa.obs['conditions'] = [i+"_"+str(j) for i,j in zip(fa.obs.Genotype, fa.obs.Age)]

# Only bacteria and interesting regions
a = fa[:, fa.var_names.isin([i for i in fa.var_names if 'Bacteria' in i])]

# Change names of some bacteria
new_gene_names_in_correct_order = ['Massilistercora-Eubacteriales incertae sedis-Eubacteriales-Clostridia-Firmicutes-Bacteria' if id=='Massilistercora-Eubacteriales-incertae-sedis-Eubacteriales-Clostridia-Firmicutes-Bacteria' else id for id in a.var_names]
new_gene_names_in_correct_order = ['Intestinimonas-Eubacteriales incertae sedis-Eubacteriales-Clostridia-Firmicutes-Bacteria' if id=='Intestinimonas-Eubacteriales-incertae-sedis-Eubacteriales-Clostridia-Firmicutes-Bacteria' else id for id in new_gene_names_in_correct_order]
a.var_names = new_gene_names_in_correct_order
bac_names = [i for i in a.var.index if i.endswith('Bacteria')]
bac_names = [i for i in bac_names if i not in ['Lacrimispora-Lachnospiraceae-Eubacteriales-Clostridia-Firmicutes-Bacteria',
                                   'Acetobacterium-Eubacteriaceae-Eubacteriales-Clostridia-Firmicutes-Bacteria']]

a = a[:, a.var_names.isin(bac_names)]
```

```python
# Select WT spots
wtatmp = a[a.obs['Genotype'] == 'WT']

# To pandas to create long format
wta = anndata.AnnData.to_df(wtatmp)
wta.reset_index(inplace=True)
wta = pd.melt(wta, id_vars=['index'], value_vars=wta.columns)

# Add annotation and spot info
annot = wtatmp.obs['annotation']
patch = wtatmp.obs['patch']
sample = wtatmp.obs['sample']
wta = wta.merge(annot, right_on=annot.index, left_on='index')
wta = wta.merge(patch, right_on=annot.index, left_on='index')
wta = wta.merge(sample, right_on=annot.index, left_on='index')
wta.drop(columns=['index'], inplace=True)
```

```python
# Plot
fig = plt.figure(figsize=(20,7))
axes = []
for row_idx in range(0,1):
    for col_idx in range(0,3):
        ax = plt.subplot2grid((1, 3), (row_idx, col_idx))
        axes.append(ax)
        
bacteria_select = ['Massilistercora-Eubacteriales incertae sedis-Eubacteriales-Clostridia-Firmicutes-Bacteria',
                  'Oscillibacter-Oscillospiraceae-Eubacteriales-Clostridia-Firmicutes-Bacteria',
                  'Pseudobutyrivibrio-Lachnospiraceae-Eubacteriales-Clostridia-Firmicutes-Bacteria']

palette = {'pellet':mcolors.to_rgba('#ce6227'),
          'mucosa and pellet':mcolors.to_rgba('#f5a791'),
          'crypt mid':mcolors.to_rgba('#58ba47'),
          'crypt base':mcolors.to_rgba('#24431d'),
          'crypt apex and mucosa':mcolors.to_rgba('#6b3b4c'),
          'crypt apex and crypt mid':mcolors.to_rgba('#829896')}

letters = ['a', 'b', 'c']

for i, bac in enumerate(bacteria_select):
    ax = axes[i]
    
    # Select bac
    wta_bac = wta[wta['variable'] == bac]

    # Plot
    y ='value'
    x = 'annotation'
    sns.stripplot(x=x, y=y, data=wta_bac, ax=ax, order=bac_regions, alpha=0.1,  s=2, palette=palette)
    sns.boxplot(x=x, y=y, data=wta_bac, ax=ax, order=bac_regions,
                palette=palette)
    
    # axis
    for tick in ax.get_xticklabels():
        tick.set_rotation(90)
        tick.set_fontsize(15)
    ax.set_xlabel('')
    ax.set_ylabel('Normalized bacterial count', fontsize=15)
    for tick in ax.get_yticklabels():
        tick.set_fontsize(15)
    
    # Titel
    ax.set_title(bac.split('-')[0], fontsize=20)
    
    ax.text(-0.15, 1,letters[i], transform=ax.transAxes, fontsize=30)
    
        
    plt.tight_layout()
    
```
