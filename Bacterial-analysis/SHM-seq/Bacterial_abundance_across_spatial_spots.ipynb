### Bacterial abundance across spatial spots

Input files:<br>
Splotch lambda GF/SPF: wtgf_lambdas_pmeans.tsv.gz<br>
Splotch information file: information.p<br>
Annotation files GF/SPF: $*$annotations.txt<br>
HE images 10015CN38_C1, 10015CN38_C2, 10015CN38_D1, 10015CN44_C1, 10015CN44_C2, 10015CN44_D2: $*$HE.jpg

```python
from __future__ import absolute_import, division, print_function

import os
import re
import glob
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pickle
import pandas as pd
from skimage import io
from skimage.color import rgb2gray
from skimage.transform import resize

import seaborn as sns

from PIL import Image
Image.MAX_IMAGE_PIXELS = 1000000000

%matplotlib inline

matplotlib.rcParams.update({'pdf.fonttype': 42,
                            'font.sans-serif': 'Arial'})
```

```python
def read_array_metadata(metadata,filename,n_levels, extraplot=None):
    array_metadata = metadata[metadata['Count file'] == filename]
    array_levels = [array_metadata['Level %d'%(idx+1)].values[0] for idx in range(0,n_levels)]
    
    if extraplot != None:
        array_levels = array_metadata['Extra_plot_level'].tolist()
    return array_levels

def resize_func(arr):
    """
    Resize image after ST well size.
    Returns x and y coordinates of the image fiited in a ST well, as well as the resized image. 
    """
    from skimage.transform import resize

    # Rescale
    image_rescaled = resize(arr, (arr.shape[0] // 10, arr.shape[1] // 10),anti_aliasing=True, preserve_range=True) #10

    xdim,ydim = image_rescaled.shape 
    testy = ydim/32-1
    testx = xdim/34-1

    x_indices = np.array([i*testy for i in range(1, 33)])
    y_indices = np.array([i*testx for i in range(1, 35)])

    x_coordinates = []
    y_coordinates = []
    for i in x_indices:
        for j in y_indices:
            x_coordinates.append(i)
            y_coordinates.append(j)

    return x_coordinates, y_coordinates, image_rescaled

def plot_spots(count_file, ax, gene, ax_count, testy, testx, sty, stx):
    coordinates = np.array([list(map(int,coordinate.split('_')))
                               for coordinate in list(lambda_posterior_means[count_file].columns)])


    cbmap = ax.scatter(testy*(coordinates[:,0]-1),
                       testx*(coordinates[:,1]-1),
                       s=15, c=lambda_posterior_means_gene[count_file].loc[gene],
                       cmap='viridis',vmin=vmin,vmax=vmax,marker='o',alpha=0.9)    
    
    ax.set_ylim(ymax = max(sty), ymin = min(sty))
    ax.set_xlim(xmax = max(stx), xmin = min(stx))

    ax.set_xticks([])
    ax.set_yticks([])

    if ax_count in [4, 9]:              
        cbar = matplotlib.pyplot.colorbar(cbmap,ax=ax,shrink=0.8)
        cbar.set_label('Normalized bacterial count', fontsize=15)
```

```python
# Load Lambda pmean df

path = ''

# Read file
filename = os.path.join(path, 'lambdas_python.tsv.gz')
    
lambda_posterior_means = pd.read_csv(filename, index_col=0, header=[0,1])
```

```python
# Change names on genes, ie remove '_'
nwe=[]
nm=lambda_posterior_means.index
for item in nm:
    nwe.append(str(item).split("_")[0])

lambda_posterior_means.index = nwe
```

```python
# Load sample_information file
path = ''
# Read file
info_file = os.path.join(path, 'information.p')

info = pickle.load(open(info_file,'rb')) 
    
metadata = info['metadata']
n_levels = info['n_levels']
```

```python
# Take exp()
lambda_posterior_means = lambda_posterior_means.astype(float)
lambda_posterior_means = np.exp(lambda_posterior_means-1)
```

# Read annotation files

```python
# Read annotation files
path = ''
annot_d = {}
region_list = []
for filename in glob.glob(os.path.join(path, '*annotations.txt')):
    
    name = filename.split('/')[-1].split('_annotations.txt')[0]
    
    if name.startswith('10015'):
    
        # Read file
        annot = pd.read_csv(filename, sep='\t')
        annot.rename(columns={'value':'annotation'}, inplace=True)

        annot_melt = pd.melt(annot, id_vars=['annotation'], value_vars=['x_y'])

        annot_melt[['x', 'y']] = annot_melt["value"].str.split("_", n = 1, expand = True)

        annot_melt[['x', 'y']] = annot_melt[['x', 'y']].astype(int)

        annot_melt = annot_melt[annot_melt['value'] != 0]

        # Get the regions
        for reg in annot_melt['annotation'].tolist():
            region_list.append(reg)
        
        reg_collasped_d = {}
        for region in np.unique(region_list):
            if region == 'pellet':
                region_collapsed = 'pellet'
            elif region == 'mucosa and pellet':
                region_collapsed = 'mucosa'
            elif region == 'crypt apex and mucosa':
                region_collapsed = 'mucosa'
            else:
                region_collapsed = 'tissue'
            reg_collasped_d[region] = region_collapsed
        
        annot_melt['annot collapsed'] = annot_melt['annotation'].map(reg_collasped_d)
        annot_d[name] = annot_melt
```

# Plot bacs

```python
# Assign colors to the regions

colors = ['blue','green','red','pink','cyan','magenta','yellow','deepskyblue', 
         'black', 'lime', 'peru', 'lightcoral', 'khaki', 'violet', 'grey',
        'fuchsia', 'lightblue', 'navy', 'darkorange', 'azure']

color_dict = {}

region_list = list(set(region_list))
region_list.sort()

for cc, reg in enumerate(region_list):
    
    color_dict[reg] = colors[cc]
```

# Individual sections

```python
# Pick a bac:

gene = 'Pseudobutyrivibrio-Lachnospiraceae-Eubacteriales-Clostridia-Firmicutes-Bacteria'

he_path = ''

lambda_posterior_means_gene = lambda_posterior_means.loc[[gene] , :]

count_files = ['/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN38_C1_stdata_adjusted.tsv',
              '/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN38_C2_stdata_adjusted.tsv',
              '/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN38_D1_stdata_adjusted.tsv',
              '/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN44_C1_stdata_adjusted.tsv',
              '/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN44_C2_stdata_adjusted.tsv',
              '/ahg/regevdata/projects/splotch_spatial_transcriptomics/britta_bac/splotch/wt_vs_gf/output6/Count_Tables/10015CN44_D2_stdata_adjusted.tsv']

vmin,vmax = 0,np.percentile(lambda_posterior_means_gene.loc[gene, (count_files, slice(None))].values,98)

fig = plt.figure(figsize=(18, 7)) 
axes = []
for ax_idx in range(0,2):
    for i in range(5):
        ax = plt.subplot2grid((2,5), (ax_idx, i))
        axes.append(ax)  
        
ax_count = 0
        
for n,count_file in enumerate(count_files):
    ax = axes[ax_count]
    name = '_'.join([count_file.split('/')[-1].split('_')[0], count_file.split('/')[-1].split('_')[1]])  
    
    if n in [0, 3]:
        ############################# Plot HE

        img_url = os.path.join(he_path, name + '_HE.jpg') 
        he_image = np.array(Image.open(img_url))

        # Get the ST scale
        stx, sty, stim = resize_func(rgb2gray(he_image))
        
        # To get xdim
        xdim,ydim = stim.shape

        testy = ydim/32-1
        testx = xdim/34-1
        
        # Plot
        ax.imshow(resize(he_image, stim.shape), aspect='auto', cmap='gray')
        ax.invert_yaxis()
        
        ax.set_xticks([])
        ax.set_yticks([])
        
        # Ylabel
        if n == 0:
            ax.set_ylabel('GF', fontsize=18)
        else:
            ax.set_ylabel('SPF', fontsize=18)

        # Title
        ax.set_title('HE: Rep 1')
            
        ############################ Plot annotation
        ax_count += 1
        ax = axes[ax_count]
        
        annot_tmp = annot_d[name]
        
        x_indices = np.array([(i-1)*testy for i in annot_tmp['x']])
        y_indices = np.array([(i-1)*testx for i in annot_tmp['y']])
                
        # add correct pixel coordiantes
        annot_tmp['new_x'] = x_indices
        annot_tmp['new_y'] = y_indices
        
        sns.scatterplot(x='new_x', y='new_y', data=annot_tmp, hue='annotation', s=15, ax=ax, palette=color_dict ) #hue='annot collapsed'

        # Put a legend to the right of the current axis
        if n == 3:
            ax.legend(loc='upper center', bbox_to_anchor=(0.5, 0), ncol=5, fontsize=14)
        else:
            ax.get_legend().remove()
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_xlabel('')
        ax.set_ylabel('')
        ax.set_ylim(ymax = max(sty), ymin = min(sty))
        ax.set_xlim(xmax = max(stx), xmin = min(stx))
        
        # Title
        ax.set_title('Annotation: Rep 1')
        
        ########################### Plot lambda
        
        ax_count += 1
        ax = axes[ax_count]
        plot_spots(count_file, ax, gene, ax_count, testy, testx, sty, stx)
        ax_count += 1
        
        # Title
        ax.set_title('Rep 1')

    ############################### Plot lambda
    else:
        ax = axes[ax_count]
        plot_spots(count_file, ax, gene, ax_count, testy, testx, sty, stx)
        
        if ax_count in [3, 8]:
            # Title
            ax.set_title('Rep 2')
        elif ax_count in [4, 9]:
            # Title
            ax.set_title('Rep 3')
            
        ax_count += 1
    
    # Main title
    plt.suptitle(gene.split('-')[0], fontsize=25)
        
fig.tight_layout()
```
