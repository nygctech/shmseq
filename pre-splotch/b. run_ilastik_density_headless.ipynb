Input files:<br> 
patches: $*$.jpg stored in sample_patches.tar.gz<br> 

```python
import os
import numpy as np
import glob
from itertools import chain
```

```python
# runs patches that have not been run before / or start from scratch
ls = []
for dirs in glob.glob(os.path.join('/patches', "*")):
    ls.append([i.split(".jpg")[0] for i in glob.glob(os.path.join(dirs, "*.jpg"))])
ls_he_patches = list(chain(*ls))

ls = []
for dirs in glob.glob(os.path.join('/patches', "*")):
    ls.append([i.split("_Probabilities.tiff")[0] for i in glob.glob(os.path.join(dirs, "*_Probabilities.tiff"))])
ls_probs_patches = list(chain(*ls))

# patches to run density probs for 
patches_to_run = np.setdiff1d(ls_he_patches, ls_probs_patches)
```
